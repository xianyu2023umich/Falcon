```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
# Define the file path
file_path = '/Users/qplazmfree/Documents/research/fortran/falcon/2_5/00001700.dat'
#file_path = '/Users/qplazmfree/Documents/research/fortran/falcon/2/test.dat'
```

```python

# Read the file
with open(file_path, 'r') as file:
    # Read the first line to get ni and nj
    ni, nj = map(int, file.readline().split())
    
    # Initialize arrays to store primitive variables
    # x = np.zeros((ni, nj))
    th = np.zeros((ni, nj))
    ph = np.zeros((ni, nj))
    var1 = np.zeros((ni, nj))
    var2 = np.zeros((ni, nj))
    var3 = np.zeros((ni, nj))
    var4 = np.zeros((ni, nj))
    var5 = np.zeros((ni, nj))
    
    # Read the rest of the file line by line
    for i in range(ni):
        for j in range(nj):
            data = list(map(float, file.readline().split()))
            # x[i, j] = data[0]
            th[i, j] = data[0]
            ph[i, j] = data[1]
            var1[i, j] = data[2]
            var2[i, j] = data[3]
            var3[i, j] = data[4]
            var4[i, j] = data[5]
            var5[i, j] = data[6]

# Function to plot each map
def plot_map(variable, title):
    plt.imshow(variable, origin='lower', cmap='viridis')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('j')
    plt.ylabel('i')
    plt.show()

data_list = [var1,var2,var3,var4,var5]  # Replace with your actual data

lats = np.linspace(-90, 90, data_list[0].shape[0])  # Define latitude values
lons = np.linspace(-180, 180, data_list[0].shape[1])  # Define longitude values
lon_grid, lat_grid = np.meshgrid(lons, lats)  # Create a meshgrid for plotting

color_maps = ['plasma', 'bwr_r', 'bwr', 'bwr', 'rainbow']
fig = plt.figure(figsize=(15, 10))

# Loop through each data map and create a subplot for each globe
for i, data in enumerate(data_list, 1):
    ax = fig.add_subplot(2, 3, i, projection=ccrs.Orthographic(central_longitude=60, central_latitude=30))
    ax.set_global()
    
    # Plot each map on its own globe
    mesh = ax.pcolormesh(lon_grid, lat_grid, data, transform=ccrs.PlateCarree(), shading='auto', cmap=color_maps[i-1])
    
    # Add individual colorbars, positioned to stay aligned with each globe
    cbar = fig.colorbar(mesh, ax=ax, orientation='horizontal', pad=0.05, aspect=30)
    cbar.set_label(f"Data Values {i+1}")

    # Optional: add title for each subplot
    ax.set_title(f"Map {i}")
    #cbar = fig.colorbar(mesh, ax=fig.get_axes(), orientation='horizontal', fraction=0.05, pad=0.07, label='Data Values')
    
    # Add coastlines if needed
    # ax.coastlines()  # Uncomment if you want coastlines

# Add a single colorbar for all maps


plt.tight_layout()
plt.show()
```
